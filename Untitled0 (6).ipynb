**1 in simple words , what are React hooks? Create a small component yhat shows a button . when the button is  clicked , toggle the text between "On" and "Off.**
--React hooks are special function in React that  allows us to use state and other React features inside functional components.
Before Hooks, state could only be used in class components . But after Hooks were introduced in React 16.8 We can now mannage  state , lifecycle method and other features inside functional componenet.

```python
import React,{useState} from 'react';

function ToggleButton(){
    const[status, setStatus]=useState('Off');
};

const handleClick=()=>{
    setStatus(status==='On'?'Off':'On');
};

return(
    <div>
    <h2>Status:{status}</h2>
    <button onClick={handleClick}>Toggle</button>
    </div>
);
}
export default ToggleButton;
```

**3  Write a componenet that fetches users from https://jsonplaceholder.typicode.com/users only once on mount and shows how many users were loaded.**

-- in react we use useEffect to perform sideEffect like fetching data.from an API.

TO makee sure the data loads only once when the components mounts, we pass an empty depency array[] to useEffect.

in given example
. we fetch users from the given API.
.Store them in state using useState.
.Display the total number of users loaded.

```python
import React,{useState,useEffect} from "react";

function UserCount(){
    const[users,setUsers]=useState([]);

    useEffect(()=>{
        fetch('https://jsonplaceholder.typicode.com/users')
        .then((response)=>response.json())
        .then((data)=>setUsers(data));
    },[]);

    return(
        <div>
        <h2>Total Users:{users.length}</h2>
        </div>
    );
}
export default UserCount;




```

**4 Create a component that starts a timer with setInterval to increase a "seconds" counter every second. stop the timer when the when the component unmount**
--In React useEffect is used to handle side effect like timers.
in above example
.we use useState to store the seconds.
.we use setInterval() to increase the second every 1 second.
. We use a cleanup function inside useEffect to stop the timer when the component unmount using clearInterval()

this prevent memory leaks and unnecessary background process.

```python
import React,{useState,useEffect} from "react";

function Timer(){
    const[seconds,setSeconds]=useState(0);
    useEffect(()=>{
        const interval=setInterval(()=>{
            setSeconds((prevSeconds)=>prevSeconds+1);
        },1000);
        return()=>clearInterval(interval);
    },[]);

    return(
        <div>
        <h2>Seconds:{seconds}</h2>
        </div>
    );
}
export default Timer;
```

**6  You have a theme value("light or Dark") this is needed by a deep child button. show how to pass it without sending theme through every parennt prop.**

--In React we can use Context API to share data between components without passing props  through every parent component.

Insted of passing theme from parent to child manually we,
.Create a Context.
.wrap component inside a provider..
.Use useContext () in the the deep child component.
this avoid "props driling".

```python
import React,{createContext,useContext} from "react";

const ThemeContext=createContext();

function App(){
    const theme='dark';

    return(
        <ThemeContext.Provider value={theme}>
        <parent/>
        </ThemeContext.Provider>
    );
}

function parent(){
    return<child/>;
}
function child(){
    const theme=useContext(ThemeContext);

    return(
        <button>Theme:{theme}</button>
    );
}
export default App;
```

**7 Make a counter with Two buttons :"+1" and"-1". Use useReduser insted of useState.**
-- In React useReduser is used to manage complex state logic.
insted of useState we use:

.A reducer function
.useReducer Hook
.Action like "increement" and "decreement"

useReducer work with:
.current state
.Action
.Return new state

```python
Import React,{useReducer} from "react";
function reducer(state,action){
    switch(action.type){
        case"increement":
            return{count:state.count+1};
        case"decreement":
            return {count:state.count-1};
        default:
            return state;
    }
}

function Counter(){
    const[state,dispatch]=useReducer(reducer,{count:0});
    return(
        <div>
        <h2>Count:{state.count}</h2>
        <button onClick={()=>dispatch({type:"increement"})}>Increment</button>
        <button onClick={()=>dispatch({type:"decreement"})}>Decrement</button>
        </div>
    );
}
export default Counter;
```

**10  Make a button that focuss an input an using a ref. Return adjacent elements without extra wrapper nodes. in one sentence say what React update process does.**

--in React

.useRef() is used tto acess a DOM  element directly.
.WE can focus the input field using ref.current.focus()
.To Return adjacent element without extra wrapper nodes, we use React Fragement
. React update process compares the virtual Dom with the real Dom and update only the changed pparts.

```python
import React,{useRef} from "react";

function FocusInput(){
    const inputRef=useRef(null);
    const focusInput=()=>{
        inputRef.current.focus();

    };
    return(
        <>
        <input type="text" ref={inputRef}/>
        <button onClick={focusInput}>Focus Input</button>
        </>
    );
}
export default FocusInput;
```
